```python
import numpy as np
from matplotlib import pyplot as plt
import os
import cv2
```

```python
# serial number list is : 22240382, 22206080, 22222961, 22240368, 22240376, 22240389

serial_number_list = ['22240382', '22206080', '22222961', '22240368', '22240376', '22240389']

image_holder = []

for serial in serial_number_list:
    image_dir = os.path.join('/home/hict/yeonkunlee/MULTIVIEW_CAL_HICT/acquired_image', serial, '0000.raw')
    fid=open(image_dir,"rb")
    Img=np.fromfile(fid,dtype='uint8', sep="")
    fid.close()
    Img=np.reshape(Img, [1536, 2048, 3])
    image_holder.append(Img)

#display image by 2 x 3 array
for i in range(6):
    # figsize = (10, 10)
    figsize = (20, 20)
    plt.subplot(2,3,i+1)
    plt.imshow(image_holder[i][:,:,::-1])
    plt.axis('off')
```

```python
total_number_converting_image = 2000

serial_number_list = ['22240382', '22206080', '22222961', '22240368', '22240376', '22240389']



for i in range(total_number_converting_image):

    for serial in serial_number_list:
        image_dir = os.path.join('/home/hict/yeonkunlee/MULTIVIEW_CAL_HICT/acquired_image', serial, '%04d.raw'%i)
        fid=open(image_dir,"rb")
        Img=np.fromfile(fid,dtype='uint8', sep="")
        fid.close()
        Img=np.reshape(Img, [1536, 2048, 3])

        # save Img as png
        save_dir = save_dir = image_dir = os.path.join('/home/hict/yeonkunlee/MULTIVIEW_CAL_HICT/acquired_image', serial+'_png')
        if not os.path.isdir(save_dir):
            os.makedirs(save_dir)
        image_dir = os.path.join(save_dir, '%04d.png'%i)
        cv2.imwrite(image_dir, Img)
    
    if i % 1000 == 0:
        print(i)


    
```
